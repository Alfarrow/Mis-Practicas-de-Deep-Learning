## Regresión lineal vs logística


|Regresión lineal|Regresión Logística|
|----------------|-------------------|
|Sirve para hacer regresión|Sive para clasificar|
|Basada en un modelo lineal|Basado en la función sigmoide|

En regresión lineal teníamos como hipótesis:
$$h_\theta(x) = \theta_0x^0 + \theta_1x^1 + \theta_2x^2,..., \theta_n x^n$$

Y ahora la hipótesis de una regresión logística está dada por:

$$h_\theta(x) = \sigma(\theta_0x^0 + \theta_1x^1 + \theta_2x^2,..., \theta_n x^n)$$

Función logística, sigmoide o sigmoidal:

$$\sigma(z) = \frac{1}{1 + e^{-(z)}}$$

¿Cuál es su mínimo y máximo?

¿Cuánto vale la ecuación cuando $z = 0, -\infty, \infty$



Graficar

```python
import numpy as np
from matplotlib import pyplot as plt

z = np.arange(-10, 10, 0.1)
y = 1/(1 + np.exp(-z))
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
plt.plot(z, np.ones_like(z) * 0.5, color = 'green', linestyle = '--', linewidth = 1)
plt.plot(z, y,color = 'orange')
```

### Entonces, ahora nuestra hipótesis puede ser descrita como:

$$0\leq h_\theta(x)\leq 1$$

Por tanto:

$$h_\theta(x) = \frac{1}{1 + e^{-Z(\theta)}}$$

siendo $Z = X.\theta$

$$h_\theta(x) = \frac{1}{1 + e^{-(\theta_0x^0 + \theta_1x^1 + \theta_2x^2,..., \theta_n x^n)}}$$

## La función de costo

### Recordemos la función de costo para regresión lineal

$$J(\theta)  = \frac{1}{2m}\sum_{i=1}^{m}(h_\theta(x_i)- y_i)^2$$

¿Porqué no usarla para regresión logística?




image.png

Como sabemos que $y$ solo puede tomar dos valores $0$ y $1$:

$$J(\theta) = \left \{
    \begin{aligned}
    &-log(h_\theta(x)), && \text{if}\ y=1 \\
    &-log(1 - h_\theta(x)), && \text{if}\ y=0
  \end{aligned}\right.$$

## Lo anterior se conoce como entropía cruzada

### Grafique la función de costo anterior:



### Para $y = 1$ 

```python
import numpy as np
from matplotlib import pyplot as plt

h = np.arange(0, 1, 0.01)
y =  -np.log(h)
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
plt.plot(h, y,color = 'orange')
```

### Para $y = 0$ 

```python
import numpy as np
from matplotlib import pyplot as plt

h = np.arange(0, 1, 0.01)
y = - np.log(1 - h)
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
plt.plot(h, y,color = 'orange')
```
