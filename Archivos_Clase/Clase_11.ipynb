La activación $a^l$ de la neurona $j$-ésima en la capa $l$-ésima está relacionada con las activaciones en la capa ($l-1$)-ésima por la ecuación:

$$a^l_j = \sigma(\sum_kw^l_{jk}a^{l-1}_k + b^l_j)$$

Escribamos la expresión en forma vectorial:

$$a^l = \sigma(w^la^{l-1} + b^l)$$

Hagamos una variable intermedia con nuestra anterior notación:

$$z^l = w^la^{l-1} + b^l$$

Por tanto:

$$a^l=\sigma(z^l)$$

Recordemos que la función costo $J$ nos da el error por tanto en backpropagation se suele sustituir la $J$ por $E$ o $C$

Para poder usar backpropagation la función de costo que utilicemos debe cumplir

$$C = \frac{1}{n}\sum_iC_i$$

Por ejemplo si usamos como función de costo el MSE (l2)
$$C = \frac{1}{2n}\sum_{i=1}^n (y_i - a^L_i)^2$$

$L$  es el número de capas
$$C_i = \frac{1}{2}(y-a^L)^2$$

Backpropagation consiste en entender cómo el cambio de los pesos y sesgos en una red cambia la función de coste. En última instancia, esto significa calcular las derivadas parciales $\frac{\partial C}{\partial w^l_{jk}}$ y 
$\frac{\partial C}{\partial b^l_j}$.

Para falicitar el cálculo de lo anterior añadimos el cálculo de $\delta^l_j$ el cuál describe el error de la neurona $j$ en la capa $l$

Es decir,

$$\delta^l_j = \frac{\partial C}{\partial z^l_j}$$

¿Porque no mejor reducimos el error $\frac{\partial C}{\partial a^l_j}$?

Respuesta: Porque es más dificil de manejar algebraícamente

El error en la capa de salida es $\delta^L$ es

 $\delta^L_j = \frac{\partial C}{\partial a^L_j}\sigma'(z^L_j)$

Lo anterior lo podemos escribir en forma vectorial como:

$$\delta^L= \nabla_aC\odot \sigma'(z^L)  $$

Aquí, $\nabla_aC$  se define como un vector cuyas componentes son las derivadas parciales $\frac{\partial C}{\partial a^L_j}$.

¿Cómo sería definido $\delta^l$ en término del error de la siguiente capa?

$$\delta^l = ((w^{l+1})^T\delta^{l+1})\odot \sigma'(z^l)$$

¿Cuáles serán las derivadas de $\frac{\partial C}{\partial w}$ y $\frac{\partial C}{\partial b}$?

Primero de la derivas en pesos y bias de una neurona en especifico

$\frac{\partial C}{\partial b^l_{j}} = \delta^l_j$

$\frac{\partial C}{\partial w^l_{jk}} = a^{l-1}_k\delta^l_j$

Dado lo anterior el algoritmo de backpropagation es:

Las ecuaciones de backpropagation nos proporcionan una forma de calcular el gradiente de la función de costo. Escribamos esto explícitamente en forma de algoritmo:

- Entrada $x$: Establece la activación correspondiente $a^1$ para la capa de entrada.
- Feedforward: Para cada $l=2,3,...,L$ calcular $z^l=w^la^{l-1}+b^l$ y $a^l=\sigma(z^l)$.
- Error de salida $\delta^L$: Calcula el vector $\delta^L= \nabla_aC\odot \sigma'(z^L)$.
- Retropropagación del error: Para cada $l=L-1,L-2,...,2$ calcular $\delta^l = ((w^{l+1})^T\delta^{l+1})\odot \sigma'(z^l)$.
- Resultado: El gradiente de la función de coste viene dado por $\frac{\partial C}{\partial w^l_{jk}} = a^{l-1}_k\delta^l_j$ y $\frac{\partial C}{\partial b^l_{j}} = \delta^l_j$.
