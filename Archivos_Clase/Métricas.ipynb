# Métricas

Apple(TP, TN, FP, FN)

```python
A_TP = 7
A_TN = 2+3+2+1
A_FP = 8+9
A_FN = 1+3
```

Orange(TP,TN,FP,FN)

```python
O_TP = 2
O_TN = 7+9+3+1
O_FP = 1+3
O_FN = 8+2
```

Mango(TP,TN,FP,FN)

```python
M_TP = 1
M_TN = 7+8+3+2
M_FP = 3+2
M_FN = 9+3
```

Funciones para Precision, Recall, Specifity, F1-score

```python
def precision(tp, fp):
  return tp/(tp+fp)

def recall(tp, fn):
  return tp/(tp+fn)

def specifity(tn, fp):
  return tn/(tn+fp)

def f1_score(precision, recall):
  return (2*precision*recall)/(precision+recall)
```

Métricas de la clase manzana

```python
P_Manzana = precision(A_TP, A_FP)
R_Manzana = recall(A_TP, A_FN)
S_Manzana = specifity(A_TN, A_FP)
F1_Manzana = f1_score(P_Manzana, R_Manzana)

print('Precisión: ', P_Manzana)
print('Recall: ', R_Manzana)
print('Specifity: ', S_Manzana)
print('F1 Score: ', F1_Manzana)
```
