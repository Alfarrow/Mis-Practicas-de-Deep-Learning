```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-10,10,.1)
sigmoide = 1/(1+np.exp(-x))
```

```python
plt.plot(x,sigmoide)
plt.grid()
plt.show()
```

```python
h = np.arange(0,1,0.01)
j1 = -np.log(h)
j2 = -np.log(1-h)
```

```python
plt.plot(h,j1,color='orange')
plt.plot(h,j2,color='red')
plt.show()
```
