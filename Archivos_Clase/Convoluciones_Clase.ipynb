```python
def apply_convolution(I, K):
  N = int(K.shape[0] / 2)
  print(N)
  I_c = np.zeros_like(I, dtype = np.float64)
  for x in range(N, I.shape[0] - N ):
    for y in range(N, I.shape[1] - N):
      for i in range(-N, N + 1):
        for j in range(-N, N + 1):
          I_c[x, y] += I[x + i, y + j] * K[ N - i, N - j]
  return I_c
```
