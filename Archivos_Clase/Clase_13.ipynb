### Hablemos un poco de métricas de rendimiento en tareas de clasificación

### Accuracy (Es la métrica más común)

Accuracy = $\frac{TP + TN}{TP + TN + FP + FN} = \frac{Aciertos}{Total de muestras}$

¿Cuáles serían los TP, TN, FP y FN?

Para Apple

- $TP$ = 7
- $TN$ = (2 + 3 + 2 + 1) = 8
- $FP$ = (8 + 9) = 17
- $FN$ = (1 + 3) = 4

###Precision

Precision = $\frac{TP}{TP+FP}$

### RECALL ó SENSITIVITY (recomendada cuando nuestras clases están desbalanceadas y la clase minoritaria es positiva)

Recall ó Sensitivity = $\frac{TP}{TP + FN}$

### SPECIFITY (recomendada cuando nuestras clases están desbalanceadas y la clase minoritaria es negativa)

Specifity = $\frac{TN}{TN + FP}$

### F1-SCORE (Es la media harmónica entre recall y precision, BUENA PARA CUANDO LAS CLASES NO ESTÁN BALANCEADAS)

F1-SCORE = $\frac{2 \times PRECISION \times RECALL}{PRECISION + RECALL}=\frac{2TP}{2TP+FP+FN}$

También se utilizan las métrica MICRO F1 y MACRO F1

- MICRO F1 se calcula mediante la suma de los TP, TN, FP y FN
- MACRO F1 se calcula como el promedio de los F1- de cada clase
- WEIGHTED F1  se cacula como la suma ponderada de los F1-SCORE multiplicados por el número de elementos de cada clase

En el caso ejemplo:

- MICRO F1 = 0.28
- MACRO F1 = 0.24
- WEIGHTED F1 = 0.24
