```python
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
X = digits.data
Y = digits.target
print(X.shape, Y.shape)
```

## Normalizar Datos

```python
print(X.max(),X.min())
```

```python
X /= 16.0
print(X.min(), X.max())
```

## Dividir Dataset

```python
from sklearn.model_selection import train_test_split#60,20,20

x_train, x_test, y_train, y_test = train_test_split(X, Y, train_size = 0.8)
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, train_size = 0.75)
```

```python
print( x_train.shape, x_val.shape, x_test.shape)
print( y_train.shape, y_val.shape, y_test.shape)
```

```python
print(y_train)
```

## Cambiando etiquetas a codificación "One Hot"

```python
from tensorflow.keras.utils import to_categorical


print(y_train[333])
y_train = to_categorical(y_train, 10)
y_val = to_categorical(y_val, 10)
y_test = to_categorical(y_test, 10)
print(y_train[333])
```

## Modo con Sequential

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense


model = Sequential()
model.add(Dense(6, input_shape=(5,), activation='sigmoid'))
model.add(Dense(6, activation='sigmoid'))
model.add(Dense(8, activation='sigmoid'))
model.add(Dense(10, activation='softmax'))
```

## Modo con Input

```python
from tensorflow.keras.layers import Input
from tensorflow.keras.models import Model
inputs = Input(shape=(64,))
x = Dense(64, activation='sigmoid')(inputs)
x = Dense(32, activation='sigmoid')(x)
x = Dense(32, activation='sigmoid')(x)
x = Dense(10, activation='softmax')(x)

model = Model(inputs=inputs, outputs=x)
```

```python
model.summary()
```

```python
import  tensorflow as tf
model.compile(loss='categorical_crossentropy', optimizer=tf.keras.optimizers.Adam(learning_rate = 1e-2), metrics = ['accuracy'])
```

```python
model.fit(x_train, y_train, validation_data = (x_val, y_val), epochs = 200 , batch_size = 32, verbose = 1)
```

## Predicciones

```python
import numpy as np

pred = model.predict(x_test) 
print(pred)
print(pred.shape)
pred = np.argmax(pred, axis = 1) 
print(pred)
print(pred.shape)
label = np.argmax(y_test,axis = 1)
```
