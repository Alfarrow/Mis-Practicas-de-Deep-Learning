## Aplicar Gradiente Descendente a (x + 3)^2

```python
import numpy as np
```

```python
def entren_grad(x, alpha, iteraciones):
  x_lista = []
  for n in range(iteraciones):
    x = x - alpha * (2*(x + 3))
    x_lista.append(x)
  return x_lista
```

```python
historial_x = entren_grad(7, 0.01, 1000)
historial_x
```

```python
from matplotlib import pyplot as plt
from IPython import display
%matplotlib inline

x = np.arange(-14, 8, 0.1)
y = (x + 3) ** 2


epoca = 1
for hx, hy in zip(historial_x, list(map(lambda x: (x + 3) ** 2, historial_x))):
  plt.plot(x, y)
  plt.scatter(hx, hy, color='r', label = 'Época: ' + str(epoca) + ', valor x: ' + str(hx))
  plt.legend(loc="upper left")
  display.clear_output(wait=True)
  plt.show()
  epoca += 1
```
